```python
from google.colab import drive
import pandas as pd
import os

# Подключаем Google Диск
drive.mount('/content/drive')

# Проверяем содержимое директории
print(os.listdir('/content/drive/MyDrive/'))

# Загружаем данные
data = pd.read_csv('/content/drive/MyDrive/Application_Data.csv')

# 5 строк
print(data.head())

```

```python
#3
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
data['Status'].hist(bins=30, edgecolor='black')
plt.title('Распределение целевой переменной')
plt.xlabel('Целевая переменная')
plt.ylabel('Частота')
plt.grid(False)
plt.show()

# Количество уникальных значений
unique_values = data['Status'].nunique()
print(f"Количество уникальных значений целевой переменной: {unique_values}")

```

2 уникальных значения (одобрено 1, не одобрено 0). Одобренных кредитов намного больше.

```python
#7
features = data.columns[:-1]
plt.figure(figsize=(20, 15))

for i, feature in enumerate(features[:5], 1):
    plt.subplot(3, 2, i)
    data[feature].hist(bins=30, edgecolor='black')
    plt.title(f'Гистограмма {feature}')
    plt.xlabel(feature)
    plt.ylabel('Частота')
    plt.grid(False)

plt.tight_layout()
plt.show()

```

```python
#8
plt.figure(figsize=(20, 15))

for i, feature in enumerate(features[:5], 1):
    plt.subplot(3, 2, i)
    plt.scatter(data[feature], data['Status'], alpha=0.5)
    plt.title(f'Зависимость {feature} от целевой переменной')
    plt.xlabel(feature)
    plt.ylabel('Целевая переменная')
    plt.grid(True)

plt.tight_layout()
plt.show()

```

```python
#9
plt.figure(figsize=(10, 6))

plt.scatter(data['Total_Bad_Debt'], data['Status'], color='blue', alpha=0.5, label='Total_Bad_Debt')
plt.scatter(data['Total_Good_Debt'], data['Status'], color='red', alpha=0.5, label='Total_Good_Debt')
plt.title('Зависимость переменных от целевой переменной')
plt.xlabel('Переменные')
plt.ylabel('Целевая переменная')
plt.legend()
plt.grid(True)
plt.show()
```
